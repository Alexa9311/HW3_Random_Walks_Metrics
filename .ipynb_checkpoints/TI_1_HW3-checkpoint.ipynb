 ### Theory on the Gradient Descent algorithm

 For the implementation the gradient descent algorithm needs define a objetive function and his gradient.
 
$$
\nabla f(x, y) = (2x, 2y)
$$

 The gradient indicates the direccion of greatest increase, so we proceed in the opposite direcction to minimize the function.

 The variables x and y are initialized and each iteration uptedate the values of:

$$ x = x - \alpha \cdot 2x
$$ $$
y = y - \alpha \cdot 2y $$

For this case, to implement the libraries numpy and matplotlib

 ### MODULES

```python
# Load modules
import numpy as np
import matplotlib.pyplot as plt
```

In this activity the objetive function is sum squares function

 ### Sum Square Funtion

$$ f(x,y) = x^2 + y^2 $$

Here is the sum of squares code

```python
# Function to be optimized
def func(x,y): 
    return x**2 + y**2 
```

 ### Code gradient of a function:

```python
# Gradient of funtion
def gradient(x,y):
    return np.array([2*x, 2*y])
```

The code implements gradient descent to minimize the funtion:
Stars from an initial point

```python
# Gradient descent algorithm
def gradient_descent(learning_rate=0.1, iterations=50, x_init=4, y_init=4):
    x,y = x_init, y_init
    history = [(x,y, func (x,y))]

    for _ in range(iterations): 
        grad = gradient (x,y)
        x -= learning_rate * grad [0]
        y -= learning_rate * grad [1]
        history.append((x,y, func(x,y)))

    return np.array(history)
```

This is the function for show the grafic in 3D

```python
# Plot funtion
def plot_3D(history):
    fig = plt.figure(figsize=(10,7))
    ax = fig.add_subplot(111, projection= '3d')
    x_vals = np.linspace(-5, 5, 100)
    y_vals = np.linspace(-5, 5, 100)
    x,y = np.meshgrid(x_vals, y_vals)
    z = func(x,y)
    
    ax.plot_surface(x,y,z, cmap = 'plasma', alpha= 0.7)
    ax.plot(history[:,0], history [:,1], history [:,2], 'ro-', markersize= 5, label = 'Trayectory')
    ax.scatter(history[0, 0], history[0,1],history[0,2], color= 'blue', label= 'star')
    ax.scatter(history[-1, 0], history[-1,1],history[-1,2], color= 'red', label= 'finish')
    
    ax.set_title("Gradient descent 3D")
    ax.set_xlabel('x')
    ax.set_ylabel('y')
    ax.set_zlabel('z')
    plt.legend()
    plt.grid(True)
    plt.show()
```

_Execution of the Gradient Descent Algorithm and ploting results_


Creates a 3D surface, showing the algorithm moves 

```python
# Run gradient descent algorithm
history = gradient_descent()
# Plot the results
plot_3D(history)
```
